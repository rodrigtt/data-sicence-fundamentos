### Fuctions


```python
def graf(x, y, color = "r--", title = "Gráfico"):
  plt.plto(x, y, color)
  plt.title(title)
  plt.grid()
```

```python
def boxplot(dado, title = ""):
  plt.boxplot(dado)
  plt.title("Boxplot " + title, fontsize = 14)
  plt.show()
```

```python
def histograma(dado, title = ""):
  plt.hist(dado)
  plt.title("Histograma " + title, fontsize = 14)
  plt.show()
```
